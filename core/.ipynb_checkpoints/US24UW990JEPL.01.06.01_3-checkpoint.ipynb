## Merge and summarize data structure.
`outer` merge. 

Rows spanned by key should have information in all its columns. 

Rows in `pdf4` which key is not in `pdf5` must have information in columns of `pdf4` but `None` in columns for `pdf5`.

Rows in `pdf5` which key is not in `pdf4` must have information in columns of `pdf5` but `None` in columns for `pdf4`.

````python
pdf = pd.merge(pdf4, pdf5, on='concat_key', how='outer', indicator=True)
pdf = pdf.where(pd.notnull(pdf), None)
````

````python
pdf.head()
````

````python
pdf.shape
````

````python
n1 = int(sum(np.isin(pdf4['concat_key'].to_numpy(), pdf5['concat_key'].to_numpy())))
n2 = int(sum(~np.isin(pdf4['concat_key'].to_numpy(), pdf5['concat_key'].to_numpy())))
n3 = int(sum(~np.isin(pdf5['concat_key'].to_numpy(), pdf4['concat_key'].to_numpy())))
n4 = pdf.shape[0]

print(n1) # Joind data.
print(n2) # Keys in f4 not in f5
print(n3) # Keys in f5 not in f4
print(n4) # Join keys, blus complement keys

# Expressed as proportions.
print(f'100*n1/n4 = {100*n1/n4}') # Proportion of joined keys (base is merged datasets with none).
print(f'100*n2/n4 = {100*n2/n4}') # Proportion of Keys in f4 not in f5.
print(f'100*n3/n4 = {100*n3/n4}') # Proportion of Keys in f5 not in f4.


assert n1 + n2 == pdf4.shape[0] and n1 + n3 == pdf5.shape[0] and n1 + n2 + n3 == n4
````
