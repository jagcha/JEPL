## Merge and summarize data structure.
`outer` merge. 

Rows spanned by key should have information in all its columns. 

Rows in `pdf4` which key is not in `pdf5` must have information in columns of `pdf4` but `None` in columns for `pdf5`.

Rows in `pdf5` which key is not in `pdf4` must have information in columns of `pdf5` but `None` in columns for `pdf4`.

````python
pdf = pd.merge(pdf4, pdf5, on='key', how='outer', indicator=True)
pdf = pdf.where(pd.notnull(pdf), None)
````

````python
pdf.head()
````

````python
pdf.tail()
````

````python
pdf.shape
````

````python
keys4 = set(pdf4['key'])
keys5 = set(pdf5['key'])

n1 = len(keys4 & keys5)  # Keys in both pdf4 and pdf5
n2 = len(keys4 - keys5)  # Keys in pdf4 but not in pdf5
n3 = len(keys5 - keys4)  # Keys in pdf5 but not in pdf4
n4 = len(keys4 | keys5)  # Total unique keys across both

# Print results
print(f'Join data: n1 = {n1}')
print(f'Keys in pdf4 not in pdf5: n2 = {n2}')
print(f'Keys in pdf5 not in pdf4: n3 = {n3}')
print(f'Join keys, plus complement keys: n4 = {n4}')

# Expressed as proportions
print(f'100*n1/n4 = {100 * n1 / n4:.2f}%')  # Proportion of joined keys
print(f'100*n2/n4 = {100 * n2 / n4:.2f}%')  # Proportion of keys in pdf4 not in pdf5
print(f'100*n3/n4 = {100 * n3 / n4:.2f}%')  # Proportion of keys in pdf5 not in pdf4

# Validate assertions
assert n1 + n2 == len(keys4)
assert n1 + n3 == len(keys5)
assert n1 + n2 + n3 == n4
````
